```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
my_img = cv2.imread("./alex.jfif",1)
my_img_gray = cv2.cvtColor(my_img, cv2.COLOR_BGR2GRAY)
```

```python
_, my_thresh_binary = cv2.threshold(my_img_gray , thresh = 128 , maxval = 255 , type=cv2.THRESH_BINARY)
```

```python
fig = plt.figure(figsize=(10,7))
rows = 1
colums = 2

fig.add_subplot(rows , colums, 1)
plt.imshow(my_img_gray , cmap= "gray")
plt.axis('off')
plt.title("original image")

fig.add_subplot(rows , colums , 2)
plt.imshow(my_thresh_binary , cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG image")
```

```python
fig = plt.figure(figsize=(12,7))
rows = 1
colums = 3

fig.add_subplot(rows , colums, 1)
plt.imshow(my_img_gray , cmap= "gray")
plt.axis('off')
plt.title("original image")

fig.add_subplot(rows , colums , 2)
plt.imshow(my_thresh_binary , cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG image")
fig.add_subplot(rows , colums , 3)

plt.imshow(my_thresh_binary_inv , cmap="gray")
plt.axis('off')
plt.title(" BINARY INV THRESHOLDÄ°NG image" )
```

```python
_, my_thresh_binary_trunc = cv2.threshold(my_img_gray , thresh = 128 , maxval = 255 , type=cv2.THRESH_TRUNC)
```

```python
fig = plt.figure(figsize=(12,7))
rows = 2
colums = 2

fig.add_subplot(rows , colums, 1)
plt.imshow(my_img_gray , cmap= "gray")
plt.axis('off')
plt.title("original image")

fig.add_subplot(rows , colums , 2)
plt.imshow(my_thresh_binary , cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG image")

fig.add_subplot(rows , colums , 3)
plt.imshow(my_thresh_binary_inv , cmap="gray")
plt.axis('off')
plt.title(" BINARY INV THRESHOLDÄ°NG image" )

fig.add_subplot(rows , colums , 4)
plt.imshow(my_thresh_binary_trunc , cmap="gray")
plt.axis('off')
plt.title(" BINARY TRUNC THRESHOLDÄ°NG image" )
```

```python
_, my_thresh_binary_tozero = cv2.threshold(my_img_gray , thresh = 128 , maxval = 255 , type=cv2.THRESH_TOZERO)
```

```python
fig = plt.figure(figsize=(15,7))
rows = 2
colums = 3

fig.add_subplot(rows , colums, 1)
plt.imshow(my_img_gray , cmap= "gray")
plt.axis('off')
plt.title("original image")

fig.add_subplot(rows , colums , 2)
plt.imshow(my_thresh_binary , cmap="gray")
plt.axis('off')
plt.title("BINARY THRESHOLDÄ°NG image")

fig.add_subplot(rows , colums , 3)
plt.imshow(my_thresh_binary_inv , cmap="gray")
plt.axis('off')
plt.title(" BINARY INV THRESHOLDÄ°NG image" )

fig.add_subplot(rows , colums , 4)
plt.imshow(my_thresh_binary_trunc , cmap="gray")
plt.axis('off')
plt.title(" BINARY TRUNC THRESHOLDÄ°NG image" )

fig.add_subplot(rows , colums , 5)
plt.imshow(my_thresh_binary_tozero , cmap="gray")
plt.axis('off')
plt.title(" BINARY TOZERO THRESHOLDÄ°NG image" )

fig.add_subplot(rows , colums , 6)
plt.imshow(my_thresh_binary_tozero_inv , cmap="gray")
plt.axis('off')
plt.title(" BINARY TOZERO INV THRESHOLDÄ°NG image" )

plt.tight_layout()
plt.savefig("./my_all_thresj_images.png",dpi=500)
plt.show()
```

```python
img_gray=cv2.imread("./camera.jfif",0)

histr =cv2.calcHist([img_gray],[0],None,[256],[0,256])

plt.plot(histr)
plt.show()
```

```python
img_gray = cv2.imread("./camera.jfif",0)
img_eq_hist = cv2.equalizeHist(img_gray)

fig = plt.figure(figsize=(10,7))
rows = 1
colums = 2

fig.add_subplot(rows , colums, 1)
plt.imshow(img_gray , cmap= "gray")
plt.axis('off')
plt.title("original image")

fig.add_subplot(rows , colums , 2)
plt.imshow(img_eq_hist , cmap="gray")
plt.axis('off')
plt.title("equalized histogram image")
```

```python
hist_org = cv2.calcHist([img_gray],[0],None,[256],[0,256])
hist_eq = cv2.calcHist([img_eq_hist],[0],None,[256],[0,256])

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(hist_org,color='b')
plt.title('Orjinal GÃ¶rÃ¼ntÃ¼ HistogramÄ±')

plt.subplot(1,2,2)
plt.plot(hist_eq,color='b')
plt.title('HistogramÄ± eÅitleme sonrasÄ± histogram')

plt.tight_layout()
plt.savefig('histogram_grfikleri.png', dpi=300)
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
img_adaptive_eq_hist = clahe.apply(img_gray)

fig = plt.figure(figsize=(10,7))
rows = 1
colums = 3

fig.add_subplot(rows , colums, 1)
plt.imshow(img_gray , cmap= "gray")
plt.axis('off')
plt.title("original image")


fig.add_subplot(rows , colums , 2)
plt.imshow(img_eq_hist , cmap="gray")
plt.axis('off')
plt.title("equalized histogram image")

fig.add_subplot(rows , colums , 3)
plt.imshow(img_adaptive_eq_hist , cmap="gray")
plt.axis('off')
plt.title(" Adaptive equalized histogram image")
```
